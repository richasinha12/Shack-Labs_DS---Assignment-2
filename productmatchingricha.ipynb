```python
import pandas as pd
import numpy as np
```

```python
amazon_data=pd.read_csv("amz_com-ecommerce_sample.csv",encoding= 'unicode_escape')
flipkart_data=pd.read_csv("flipkart_com-ecommerce_sample.csv",encoding= 'unicode_escape')
flipkart_data.head()
```

```python
amazon_data.head()
```

```python
amazon_data.isnull().sum()
```

```python
flipkart_data.isnull().sum()
```

```python
req_amazon = amazon_data[["product_name",'retail_price','discounted_price']]
req_flipkart = flipkart_data[["product_name",'retail_price','discounted_price']]

df2=req_amazon.compare(req_flipkart,keep_shape=True, keep_equal=True)
```

```python
df2.rename(columns={"self":"Product Name Amazon"})

# product_name=.rename(columns={"self":"Product Name in Amazon","other":"Product Name in Flipkart"})
# discounted=.rename(columns={"self":"Discounted Price in Amazon","other":"Discounted Price in Flipkart"})
# retail=.rename(columns={"self":"Retail Price in Amazon","other":"Retail Price in Flipkart"})

solution =  pd.concat([df2["product_name"],df2["retail_price"],df2["discounted_price"]],axis=1,ignore_index=True)
solution.columns = ["Product Name in Amazon","Product Name in Flipkart","Retail Price in Amazon","Retail Price in Flipkart","Discounted Price in Amazon","Discounted Price in Flipkart"]
solution=solution.iloc[:,[1,3,5,0,2,4]]
solution.head()
```

```python
solution.to_csv("Flipkart_Amazon_price_compare.csv")
```
